# Data Sourcing: Getting ACS data using the census API

```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from census import Census
from us import states
import numpy as np
import os
```

```python
#Initialize Census object using census API key
census_object = Census("bf6690c63fb4bbd43ccae839241e9be45bfc0881")
```

```python
# Obtain population for every county in all states
pop_all_counties = census_object.acs5.state_county(fields = ('NAME', 'B01003_001E'),
                                      state_fips = "*",
                                      county_fips = "*",
                                      year = 2017)
```

```python
#Convert API pull into a dataframe
pop_all_counties_df = pd.DataFrame(pop_all_counties)
pop_all_counties_df
```

```python
# Download shapefile of US Census tracts
counties = gpd.read_file("https://www2.census.gov/geo/tiger/GENZ2021/shp/cb_2021_us_county_500k.zip")

#Convert projection system
#counties = counties.to_crs(epsg = 32617)

counties.head(2)
```

```python
#Print out data types for two data frames
print("Column data types for census data:\n{}".format(pop_all_counties_df.dtypes))

# Check column data types for census shapefile
print("\nColumn data types for census shapefile:\n{}".format(counties.dtypes))
```

```python
#Join together

#Create primary key to match shapefile
pop_all_counties_df['GEOID'] = pop_all_counties_df['state'] + pop_all_counties_df['county']

# Join the attributes of the dataframes together
# Source: https://geopandas.org/docs/user_guide/mergingdata.html
merged = pop_all_counties_df.merge(counties, on = "GEOID")
merged = gpd.GeoDataFrame(merged)

# Show result
merged.head(2)
```

```python
merged.head()
```

```python
# Create subplots
fig, ax = plt.subplots(1, 1, figsize = (20,10))

# Plot data
# Source: https://geopandas.readthedocs.io/en/latest/docs/user_guide/mapping.html
states48 = merged[ merged['state'] != '15']
states48 = states48[ states48['state'] != '02']
states48[['geometry','B01003_001E']].plot(column = "B01003_001E",
                       ax = ax,
                        kind='geo',
                       cmap = "RdPu",
                       legend = True)

# Stylize plots
plt.style.use('bmh')

# Set title
ax.set_title('Population by County', fontdict = {'fontsize': '25', 'fontweight' : '3'})
```
